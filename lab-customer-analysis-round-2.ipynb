```python
# Show the dataframe shape.
import pandas as pd
import numpy as np
df = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
df.shape
```

```python
# Standardize header names.
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower())
df.columns = cols
```

```python
# Which columns are numerical?
df._get_numeric_data()
```

```python
# Which columns are categorical?
df.select_dtypes('object')
```

```python
#Check and deal with NaN values.
df.isna().sum() 
```

```python
round(df.isna().sum()/len(df),4)*100
```

```python
nulls_df = pd.DataFrame(round(df.isna().sum()/len(df),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
df
```

```python
columns_drop = nulls_df[nulls_df['percent_nulls']>50]['header_name'] 
print(columns_drop.values)
```

```python
df = df.drop(['vehicle type'],axis=1)
```

```python
# to replace numerical NAN values
mean_months_since_last_claim = np.mean(df['months since last claim'])
df['months since last claim'] = df['months since last claim'].fillna(mean_months_since_last_claim)
mean_number_of_open_complaints = np.mean(df['number of open complaints'])
df['number of open complaints'] = df['number of open complaints'].fillna(mean_number_of_open_complaints)
```

```python
df['state'].value_counts()
```

```python
# to replace categorical NANs 
# create a random permuation of the categorical values
permutation = np.random.permutation(df['state'])

#erase the empty values
empty_is = np.where(permutation == "")
permutation = np.delete(permutation, empty_is)

#replace all empty values of the dataframe[field]
end = len(permutation)
df['state'] = df['state'].apply(lambda x: permutation[np.random.randint(end)] if pd.isnull(x) else x)
```

```python
df['state'].value_counts()
```

```python
# repeat for 'response'
permutation = np.random.permutation(df['response'])
empty_is = np.where(permutation == "")
permutation = np.delete(permutation, empty_is)
end = len(permutation)
df['response'] = df['response'].apply(lambda x: permutation[np.random.randint(end)] if pd.isnull(x) else x)
```

```python
# repeat for 'vehicle class'
permutation = np.random.permutation(df['vehicle class'])
empty_is = np.where(permutation == "")
permutation = np.delete(permutation, empty_is)
end = len(permutation)
df['vehicle class'] = df['vehicle class'].apply(lambda x: permutation[np.random.randint(end)] if pd.isnull(x) else x)
```

```python
# repeat for 'vehicle size'
permutation = np.random.permutation(df['vehicle size'])
empty_is = np.where(permutation == "")
permutation = np.delete(permutation, empty_is)
end = len(permutation)
df['vehicle size'] = df['vehicle size'].apply(lambda x: permutation[np.random.randint(end)] if pd.isnull(x) else x)
```

```python
# check number of NANs
df.isna().sum() 
```

```python
 #Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.
df['effective to date'] = pd.to_datetime(df['effective to date'], errors='coerce')
```
